# Notebook de Análisis de Datos

Este notebook está diseñado para análisis de datos y ciencia de datos.

```python
# Importar librerías necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de visualización
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
%matplotlib inline
```

## 1. Carga de Datos

Comienza cargando tus datos aquí.

```python
# Ejemplo: Cargar datos desde un archivo CSV
# df = pd.read_csv('ruta/a/tu/archivo.csv')

# Crear un DataFrame de ejemplo
df = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.randn(100),
    'C': np.random.randint(0, 10, 100)
})

df.head()
```

## 2. Exploración de Datos

Explora las características básicas de tus datos.

```python
# Información general del dataset
print("Dimensiones:", df.shape)
print("\nInformación:")
df.info()
print("\nEstadísticas descriptivas:")
df.describe()
```

## 3. Visualización de Datos

Crea visualizaciones para entender mejor tus datos.

```python
# Ejemplo de visualización
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Histograma
df['A'].hist(bins=30, ax=axes[0], edgecolor='black')
axes[0].set_title('Distribución de A')
axes[0].set_xlabel('Valor')
axes[0].set_ylabel('Frecuencia')

# Scatter plot
axes[1].scatter(df['A'], df['B'], alpha=0.5)
axes[1].set_title('Relación entre A y B')
axes[1].set_xlabel('A')
axes[1].set_ylabel('B')

plt.tight_layout()
plt.show()
```
