Al realizar una consulta en un registro hemos obtenido una cadena de texto corrupta al revés.
Al parecer contiene el nombre de un alumno y la nota de un exámen.
¿Cómo podríamos formatear la cadena y conseguir una estructura como la siguiente?:

* ***Nombre*** ***Apellido*** ha sacado un ***Nota*** de nota.

*Ayuda: Para voltear una cadena rápidamente utilizando slicing podemos utilizar un tercer índice -1: cadena[::-1]*

````python
cadena = "zeréP epeP,81"

# Completa el ejercicio aquí

cadena = "zeréP epeP,81"

# Voltear la cadena
cadena_volteada = cadena[::-1]  # "18,Pepe Pérez"

# Separar nota y nombre
nota = cadena_volteada.split(",")[0]           # "18"
nombre_apellido = cadena_volteada.split(",")[1]  # "Pepe Pérez"

print(f"{nombre_apellido} ha sacado un {nota} de nota.")
````

**Ejercicio 3**

Genere una lista de números enteros del 1 al 99 (usando *list* y *range*) y elimine de la lista los múltiplos de 5 y de 7. Finalmente muestre el listado los números restantes.

````python
# Completa el ejercicio aquí

numeros = list(range(1, 100))
numeros = [n for n in numeros if n % 5 != 0 and n % 7 != 0]
print(numeros)
````

**Ejercicio 4**

Crear los siguientes conjuntos: {a,b,c,d,e,f,g,h,i} y {a,e,i,o,u}, y realice las operaciones de unión, intersección y diferencia de conjuntos. Repita lo anterior con estos conjuntos: {1,2,3,4,5,6,7,8,9} y {1,3,5,7,9}

````python
# Completa el ejercicio aquí

# Conjuntos de letras
conjunto1 = {'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i'}
conjunto2 = {'a', 'e', 'i', 'o', 'u'}

print(f"Unión: {conjunto1 | conjunto2}")
print(f"Intersección: {conjunto1 & conjunto2}")
print(f"Diferencia (1-2): {conjunto1 - conjunto2}")

# Conjuntos de números
conjunto3 = {1, 2, 3, 4, 5, 6, 7, 8, 9}
conjunto4 = {1, 3, 5, 7, 9}

print(f"Unión: {conjunto3 | conjunto4}")
print(f"Intersección: {conjunto3 & conjunto4}")
print(f"Diferencia (3-4): {conjunto3 - conjunto4}")
````

**Ejercicio 5**

Usando conjuntos y listas, ingrese algunas líneas de código que permita ingresar una palabra, convertirlo en mayúscula, que elimine las vocales y queden las consonantes distintas. Finalmente debe devolver el listado de consonantes en orden descendente.

* Ingresa una palabra: murciélago
* ['R', 'M', 'L', 'G', 'C']

````python
# Completa el ejercicio aquí

palabra = input("Ingresa una palabra: ")
palabra_mayus = palabra.upper()

vocales = {'A', 'E', 'I', 'O', 'U'}
consonantes = set(palabra_mayus) - vocales

consonantes_ordenadas = sorted(list(consonantes), reverse=True)
print(consonantes_ordenadas)
````

**Ejercicio 6**

Almacene en un diccionario los precios de las frutas de la siguiente tabla. Luego pregunte al usuario por una fruta, el número de kilos y muestre por pantalla el precio de ese número de kilos de fruta. Si la fruta no está en el diccionario debe mostrar un mensaje informando de ello.


````python
# Completa el ejercicio aquí

precios = {
    "Plátano": 1.35,
    "Manzana": 0.80,
    "Pera": 0.85,
    "Naranja": 0.70
}

fruta = input("Ingrese una fruta: ")
kilos = float(input("Ingrese el número de kilos: "))

if fruta in precios:
    total = precios[fruta] * kilos
    print(f"El precio de {kilos} kilos de {fruta} es {total}")
else:
    print("La fruta no está en el diccionario")
````

**Ejercicio 7**

Un asistente digital administra una lista con todas las tareas pendientes del día.
El sistema debe permitir:

Agregar varias tareas nuevas,

Eliminar una tarea que ya se completó,

Quitar la última tarea añadida en caso de error,

Ordenar las tareas alfabéticamente,

Invertir la lista completa para mostrar prioridades desde la última agregada.


**Tareas = ["Pagar servicios", "Revisar correos", "Llamar al proveedor"]**

````python
# Completa el ejercicio aquí

Tareas = ["Pagar servicios", "Revisar correos", "Llamar al proveedor"]

# Agregar varias tareas nuevas
Tareas.append("Enviar informe")
Tareas.append("Comprar suministros")
Tareas.append("Agendar reunión")

# Eliminar una tarea que ya se completó
Tareas.remove("Revisar correos")

# Quitar la última tarea añadida en caso de error
Tareas.pop()

# Ordenar las tareas alfabéticamente
Tareas.sort()

# Invertir la lista para mostrar prioridades desde la última agregada
Tareas.reverse()

print(f"Tareas pendientes: {Tareas}")
````
