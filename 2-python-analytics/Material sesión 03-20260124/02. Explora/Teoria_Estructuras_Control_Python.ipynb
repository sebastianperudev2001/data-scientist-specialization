## Estructuras Condicionales en Python


### Introducción a Estructuras Condicionales
Las estructuras condicionales permiten que un programa tome decisiones en función de condiciones específicas. En Python, las decisiones se toman utilizando las declaraciones `if`, `elif` y `else`.


### Sintaxis de `if`, `elif`, `else`

```python
x = 35
x
```

```python
x < 18
```

```python
if x < 18:
    print("menor de edad")
else:
    print("Mayor de edad")
```

```python

# Ejemplo básico de if, elif, else
x = int(input("Ingresa un número: "))

if x > 0:
    print("x es positivo")
elif x == 0:
    print("x es cero")
else:
    print("x es negativo")
```

```python
x = int(input("Ingresa una edad: "))

if x < 18:
    print("menor de edad")
elif x < 35:
    print("joven trabajador")
elif x < 65:
    print("Aduto")
else:
    print("jubilado")
```


- `if` evalúa una condición; si es verdadera, ejecuta el bloque de código que sigue.
- `elif` (else if) evalúa una nueva condición si la anterior fue falsa.
- `else` se ejecuta si todas las condiciones anteriores son falsas.

## Estructuras Iterativas en Python


### Introducción a Estructuras Iterativas
Las estructuras iterativas permiten repetir un bloque de código varias veces. Python ofrece dos estructuras iterativas principales: `for` y `while`.


### Estructura `for`

```python
dias = ["lunes", "martes","miercoles","jueves", "viernes","sabado","domingo"]

for dia in dias:
    print(dia)
```

```python
list(range(5))
```

```python

# Ejemplo básico de for
for i in range(5):
    print("Iteración", i)
```

```python
colores = ["azul", "rojo", "verde", "amarillo", "negro"]
colores
```

```python
for color in colores:
    print("el color es:", color)
```

```python
list(range(len(colores)))
```

```python
for i in range(len(colores)):
    print("el color es", colores[i], "el indice es", i)
```

```python
tupla = (1, "Hola", "Chau", "Tengo hambre", True)

for i in tupla:
    print(i)
```


- `for` se utiliza para iterar sobre una secuencia (como una lista, tupla, diccionario o rango).
- En el ejemplo anterior, `range(5)` genera una secuencia de números del 0 al 4, y `i` toma cada uno de esos valores en cada iteración.


### Bucle for sobre un string (letra por letra)

```python
for letter in "Sunday":
    print(letter)
```

```python
# Letra por letra de forma invertida
day = "Sunday"
day_inv = day[::-1]   # invierte el orden de las letras de un string
for letter in day_inv:
    print(letter)
```

### Estructura `while`

```python

# Ejemplo básico de while
contador = 0

while contador < 5:
    print("Contador:", contador)
    contador += 1 # contador = contador + 1
```

```python
anio = 2008

while anio < 2024:
    print("el reporte del", anio)
    anio += 2 # anio = anio + 2
```


- `while` repite un bloque de código mientras una condición sea verdadera.
- En el ejemplo anterior, el ciclo `while` continúa ejecutándose mientras `contador` sea menor que 5.


### Uso de `break` y `continue`

```python
list(range(10))
```

```python
# Ejemplo de uso de break
for numero in range(10):
    if numero == 5:
        break  # Salir del ciclo cuando numero es 5
    print(numero)
```

```python
# Ejemplo de uso de continue
for numero in range(10):
    if numero % 2 == 0:
        continue  # Saltar a la siguiente iteración si numero es par
    print(numero)
```

```python

# Ejemplo de uso de break
for numero in range(10):
    if numero == 5:
        break  # Salir del ciclo cuando numero es 5
    print(numero)

# Ejemplo de uso de continue
for numero in range(10):
    if numero % 2 == 0:
        continue  # Saltar a la siguiente iteración si numero es par
    print(numero)
```


- `break` se utiliza para salir de un bucle antes de que haya terminado normalmente.
- `continue` se utiliza para saltar el resto del código dentro del bucle y pasar a la siguiente iteración.


## Ejemplo Avanzado: Combinando `if`, `for` y `while`

```python

# Ejemplo avanzado
n = 10
suma = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        suma += i

while suma > 0:
    print("Suma es positiva:", suma)
    suma -= 5

print("Bucle terminado.")
# print(suma)
```
