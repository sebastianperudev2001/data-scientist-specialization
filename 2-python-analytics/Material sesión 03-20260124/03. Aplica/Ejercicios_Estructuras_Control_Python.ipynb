# Ejercicios: Estructuras de Control Iterativas y Condicionales en Python

## Ejercicio 1: Uso de `if`, `elif`, `else`


1. **Escribe un programa que solicite al usuario un número y determine si es positivo, negativo o cero. Imprime un mensaje correspondiente.**
2. **Modifica el programa para que también indique si el número es par o impar.**


## Ejercicio 2: Uso de `for`


1. **Escribe un programa que imprima los números del 1 al 10 utilizando un bucle `for`.**
2. **Modifica el programa para que también imprima el cuadrado de cada número.**
3. **Escribe un programa que recorra una lista de nombres e imprima un saludo para cada nombre.**


## Ejercicio 3: Uso de `while`


1. **Escribe un programa que solicite al usuario un número positivo y continúe solicitando números hasta que el usuario ingrese un número negativo.**
2. **Modifica el programa para que, al final, imprima la suma de todos los números positivos ingresados.**


## Ejercicio 4: Uso de `break` y `continue`


1. **Escribe un programa que imprima los números del 1 al 10. Utiliza `break` para salir del bucle si el número es igual a 5.**
2. **Escribe un programa que imprima los números del 1 al 10, pero utiliza `continue` para saltar la impresión si el número es divisible por 3.**
