# Ejercicios: Listas y Tuplas en Python

## Ejercicio 1: Creación y Manipulación de Listas


1. Crea una lista llamada `mi_lista` que contenga los siguientes elementos: `10`, `20`, `30`, `40`, `50`.
2. Accede al primer y último elemento de la lista. Imprime ambos elementos.
3. Modifica el tercer elemento de la lista para que su nuevo valor sea `35`. Imprime la lista completa.
4. Añade el número `60` al final de la lista. Imprime la lista actualizada.
5. Elimina el segundo elemento de la lista. Imprime la lista actualizada.
6. Ordena la lista en orden descendente. Imprime la lista ordenada.


## Ejercicio 2: Indexing y Slicing en Listas


1. **Dada la lista `nueva_lista = [5, 10, 15, 20, 25, 30, 35, 40]`, accede al cuarto elemento e imprímelo.**
2. **Utiliza slicing para obtener los primeros 4 elementos de la lista `nueva_lista`. Imprime el resultado.**
3. **Utiliza slicing para obtener los últimos 3 elementos de la lista `nueva_lista`. Imprime el resultado.**
4. **Crea una sublista que contenga los elementos desde el segundo hasta el quinto (inclusive) de `nueva_lista`. Imprime la sublista.**


## Ejercicio 3: Creación y Acceso a Elementos en Tuplas


1. **Crea una tupla llamada `mi_tupla` con los siguientes elementos:** `100`, `"Python"`, `3.14`, `True`.
2. **Accede al segundo elemento de `mi_tupla` e imprímelo.**
3. **Intenta modificar el primer elemento de la tupla `mi_tupla` a `200`. ¿Qué sucede?**
4. **Accede al último elemento de la tupla utilizando un índice negativo e imprímelo.**


## Ejercicio 4: Comparación entre Listas y Tuplas


1. **Crea una lista llamada `mi_lista_comp` con los siguientes elementos:** `1`, `2`, `3`, `4`.
2. **Crea una tupla llamada `mi_tupla_comp` con los mismos elementos:** `1`, `2`, `3`, `4`.
3. **Intenta cambiar el tercer elemento de la lista `mi_lista_comp` a `10`. Imprime la lista.**
4. **Intenta hacer lo mismo con la tupla `mi_tupla_comp`. ¿Qué sucede?**
5. **¿Qué conclusiones puedes sacar sobre la mutabilidad de listas y tuplas?**
