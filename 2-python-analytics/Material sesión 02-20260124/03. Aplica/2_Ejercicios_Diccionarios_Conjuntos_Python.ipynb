# Ejercicios: Diccionarios y Conjuntos en Python

## Ejercicio 1: Creación y Manipulación de Diccionarios


1. **Crea un diccionario llamado `inventario` que contenga los siguientes pares clave-valor:** `"manzanas": 50`, `"naranjas": 30`, `"plátanos": 20`.
2. **Accede al valor asociado con la clave `"naranjas"` y imprímelo.**
3. **Modifica el valor de `"plátanos"` para que sea `25`. Imprime el diccionario actualizado.**
4. **Añade un nuevo par clave-valor al diccionario: `"uvas": 100`. Imprime el diccionario actualizado.**
5. **Elimina el par clave-valor correspondiente a `"manzanas"` del diccionario. Imprime el diccionario actualizado.**


## Ejercicio 2: Métodos Comunes de Diccionarios


1. **Utiliza el diccionario `inventario` creado en el Ejercicio 1.**
2. **Obtén una lista de todas las claves en el diccionario e imprímela.**
3. **Obtén una lista de todos los valores en el diccionario e imprímela.**
4. **Obtén una lista de todos los pares clave-valor en el diccionario e imprímela.**
5. **Usa el método `get()` para obtener el valor asociado con la clave `"uvas"`.**
6. **Usa el método `pop()` para eliminar el par clave-valor asociado con `"naranjas"` y devuelve su valor. Imprime el valor devuelto y el diccionario actualizado.**


## Ejercicio 3: Creación y Manipulación de Conjuntos


1. **Crea un conjunto llamado `conjunto_a` con los elementos:** `1`, `2`, `3`, `4`, `5`.
2. **Crea un segundo conjunto llamado `conjunto_b` con los elementos:** `4`, `5`, `6`, `7`, `8`.
3. **Añade el elemento `9` a `conjunto_a`. Imprime el conjunto actualizado.**
4. **Elimina el elemento `2` de `conjunto_a`. Imprime el conjunto actualizado.**
5. **Encuentra la unión de `conjunto_a` y `conjunto_b` e imprímela.**
6. **Encuentra la intersección de `conjunto_a` y `conjunto_b` e imprímela.**
7. **Encuentra la diferencia entre `conjunto_a` y `conjunto_b` e imprímela.**


## Ejercicio 4: Comparación entre Diccionarios y Conjuntos


1. **Crea un diccionario llamado `productos` con las claves `"manzanas"`, `"naranjas"` y `"plátanos"` con valores `5`, `10` y `7` respectivamente.**
2. **Crea un conjunto llamado `frutas` con los elementos `"manzanas"`, `"uvas"` y `"naranjas"`.**
3. **Usa un bucle para recorrer las claves del diccionario `productos` y verifica si cada clave está presente en el conjunto `frutas`. Imprime un mensaje indicando si la clave está o no en el conjunto.**
